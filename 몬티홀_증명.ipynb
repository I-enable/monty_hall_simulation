```python
import random

idx = 0


# 카운트 변수
car_cnt = 0
no_car_cnt = 0
car_cnt_z = 0

# 카운터 리스트
cnt_list = []
idx_list = list(range(1,10001))

# 몬티홀 문
door = ["goat_one", "goat_two", "car"]

# 문을 오픈한 경우
open_door_one = door[0]
open_door_two = door[1]
open_door_thr = door[2]

# 바꿀 경우
while idx < 10000:
    choice = random.choice(door)
    
    if choice == "goat_one":
        second = open_door_two
    elif choice == "goat_two":
        second = open_door_one
    elif choice == "car":
        second = open_door_two
    
    # 2번 선택 1번 개방
    if second == open_door_one:
        
        # 선택을 바꿀경우
        car_cnt += 1

    
    # 3번 선택 2번 개방
    elif second == open_door_two and choice == "car":
        
        # 선택을 바꿀경우
        no_car_cnt += 1
    
    # 1번 선택 2번 개방
    elif second == open_door_two and choice == "goat_one":
        
        # 선택을 바꿀경우
        car_cnt += 1
        # 사회자 전략확률 추가
        car_cnt_z += 1
        
        
    # 시행횟수
    idx += 1
    
    cnt_list.append((car_cnt/idx)*100)

print("자동차를 고를 확률 :",(car_cnt/idx)*100,"%")
print("사회자 변동확률 :",((car_cnt_z + no_car_cnt)/idx)*100,"%")
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(25,5))
plt.plot(idx_list, cnt_list)
plt.title("Monty_Hall")
plt.xlabel("idx")
plt.ylabel("percent")
plt.show()
```
